### Spatial Correlation Script

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from nilearn import maskers
from nilearn import image
from nilearn.image import load_img
from nilearn.image import math_img
import nibabel as nib
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mask = load_img('/Users/jp1590/Desktop/MNI152_T1_0.5mm.nii.gz')
mask_bin = math_img('img > 0', img=mask)
nib.save(mask_bin,'/Users/jp1590/Desktop/Binary_MNI152_T1_0.5mm.nii.gz')
```

```python
masker = maskers.NiftiMasker(load_img('/Users/jp1590/Desktop/Binary_MNI152_T1_0.5mm.nii.gz')).fit()
def single_corr(img1, img2):
    vec1 = masker.transform(img1)[0,:]
    vec2 = masker.transform(img2)[0,:]
    return pearsonr(vec1, vec2)[0]
```

```python
img1 = load_img('/Volumes/Cingulate/Fred/Maastricht_n30/Outputs/Crop_NoFlip/andy_0.1_34%_pearson_noflip.nii.gz')
img2 = load_img('/Volumes/Cingulate/Fred/Maastricht_n30/Outputs/Crop_NoFlip/garance_0.1_30%_pearson_noflip.nii.gz')
print(single_corr(img1,img2))
```

```python
rmap_list = ['andy_0.1_34%_pearson_noflip.nii.gz','andy_0.1_34%_spearman_noflip.nii.gz','garance_0.1_30%_pearson_noflip.nii.gz','garance_0.1_30%_spearman_noflip.nii.gz','mine_0.2_50%_spearman_noflip.nii.gz','mine_0.2_50%_pearson_noflip.nii.gz']
r_vals = {r: [] for r in rmap_list}

def r_corr(constant,out_list):
    img1 = load_img(f'/Volumes/Cingulate/Fred/Maastricht_n30/Outputs/Crop_NoFlip/{constant}')
    for i in rmap_list:
        img2 = load_img(f'/Volumes/Cingulate/Fred/Maastricht_n30/Outputs/Crop_NoFlip/{i}')
        out_list.append(single_corr(img1, img2))
        
for k in range(0,len(rmap_list)):
    r_corr(rmap_list[k],r_vals[rmap_list[k]])
```

```python
r_values_df = pd.DataFrame.from_dict(r_vals, orient='index').transpose()
r_values_df.to_csv('/Volumes/Cingulate/Fred/Maastricht_n30/CSV/maastricht_corrmat.csv')
```

```python
df = pd.read_csv('/Volumes/Cingulate/Fred/Maastricht_n30/CSV/maastricht_corrmat.csv')
df.set_index('First_Corr_Var', inplace=True) #so the first column is not plotted, set as titles
plt.figure(figsize=(16, 6))
ax = sns.heatmap(df, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Maastricht R-Maps')
plt.xlabel('First R-Map')
plt.ylabel('Second R-Map')
plt.show()
```
